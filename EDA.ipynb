## Data Cleaning

```python
clean_df = users.copy()
```

```python
# fix object type columns

def to_dict_or_none(x):
    if isinstance(x, dict):
        return x
    if isinstance(x, str) and x.strip().startswith('{'):
        try:
            return ast.literal_eval(x)
        except Exception:
            return None
    return None

for col in ['hair', 'address', 'bank', 'company', 'crypto']:
    clean_df[col] = clean_df[col].apply(to_dict_or_none)

hair_df = pd.json_normalize(clean_df['hair']).add_prefix('hair_')
clean_df = pd.concat([clean_df.drop(columns=['hair']), hair_df], axis=1)

address_df = pd.json_normalize(clean_df['address'])
clean_df = pd.concat([clean_df.drop(columns=['address']), address_df], axis=1)

bank_df = pd.json_normalize(clean_df['bank']).add_prefix('bank_')
clean_df = pd.concat([clean_df.drop(columns=['bank']), bank_df], axis=1)

company_df = pd.json_normalize(clean_df['company']).add_prefix('company_')
clean_df = pd.concat([clean_df.drop(columns=['company']), company_df], axis=1)

crypto_df = pd.json_normalize(clean_df['crypto']).add_prefix('crypto_')
clean_df = pd.concat([clean_df.drop(columns=['crypto']), crypto_df], axis=1)

print(f"Columns after expansion: {clean_df.shape}")
```

```python
clean_df['birthDate'] = pd.to_datetime(clean_df['birthDate'], errors='coerce')
```

```python
sensitive_cols = ['password', 'ssn', 'ein', 'ip']
clean_df = clean_df.drop(columns=sensitive_cols)
```

```python
clean_df['maidenName'] = clean_df['maidenName'].fillna('Unknown')
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
age_bins = [20, 25, 30, 35, 40, 50]
age_labels = ['21-25', '26-30', '31-35', '36-40', '41+']
clean_df['age_group'] = pd.cut(clean_df['age'], bins=age_bins, labels=age_labels, include_lowest=True)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

clean_df['role'].value_counts().plot(kind='bar', ax=axes[0, 0])
axes[0, 0].set_title('Role Distribution')
axes[0, 0].set_xlabel('Role')
axes[0, 0].set_ylabel('Count')

clean_df['age'].plot(kind='hist', bins=10, ax=axes[0, 1], edgecolor='black')
axes[0, 1].set_title('Age Distribution')
axes[0, 1].set_xlabel('Age')
axes[0, 1].set_ylabel('Frequency')

clean_df['age_group'].value_counts().sort_index().plot(kind='bar', ax=axes[1, 0])
axes[1, 0].set_title('Age Group Distribution')
axes[1, 0].set_xlabel('Age Group')
axes[1, 0].set_ylabel('Count')

clean_df['bloodGroup'].value_counts().head(6).plot(kind='bar', ax=axes[1, 1])
axes[1, 1].set_title('Top Blood Groups')
axes[1, 1].set_xlabel('Blood Group')
axes[1, 1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
top5_cities = clean_df['city'].value_counts().head(5).index
city_gender = (
    clean_df[clean_df['city'].isin(top5_cities)]
    .groupby(['city', 'gender'])
    .size()
    .unstack(fill_value=0)
    .loc[top5_cities]
)

ax = city_gender.plot(kind='bar', figsize=(10, 5), color=["#F371BB", "#6DC972"])
ax.set_title('Gender Distribution in Top 5 Cities')
ax.set_xlabel('City')
ax.set_ylabel('Count')
ax.legend(title='Gender')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()

city_gender
```

```python
clean_df.company_department.value_counts().plot(kind='bar', figsize=(10, 5), color="#6DC972")
plt.title('Company Department Distribution')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
```

```python
dept_counts = clean_df['company_department'].value_counts().head(8)

dept_gender = (
    clean_df[clean_df['company_department'].isin(dept_counts.index)]
    .groupby(['company_department', 'gender'])
    .size()
    .unstack(fill_value=0)
    .loc[dept_counts.index]
)

ax = dept_gender.plot(kind='bar', stacked=True, figsize=(11, 5), color=["#F371BB", "#6DC972"])
ax.set_title('Gender Distribution by Department (Top 8)')
ax.set_xlabel('Department')
ax.set_ylabel('Count')
ax.tick_params(axis='x', rotation=35)
ax.legend(title='Gender')
plt.tight_layout()
plt.show()

dept_gender
```
